```python
# my answer
img = cv2.imread("imori.jpg")
shape = img.shape
print(shape[0]//2, shape[0]/2)


B = img[:int(shape[0]/2), :int(shape[1]/2),0].copy()

img[:int(shape[0]/2), :int(shape[1]/2),0] = img[:int(shape[0]/2), :int(shape[1]/2),2]
img[:int(shape[0]/2), :int(shape[1]/2),2] = B

plt.imshow(img)
```
